## Web Scraping News Articles   

```python
import os 
import json
import pandas as pd
from time import time
```

```python

def list_files(file_path):
    for root, dir, file in os.walk(file_path):
        for files in file:
            yield os.path.join(root,files)
```

```python
json_list = []
dir_path = "/Users/khushal/Desktop/Projects/news_archive"
file_paths = list(list_files(dir_path))

start_time = time()

for item in file_paths:
    if item.endswith('.json'):
        with open(item) as f:
            for line in f:
                data = json.loads(line)
                json_list.append([data['published'], data['thread']['site'], data['title'], data['text'], data['url']])

end_time = time()
execution_time = end_time - start_time

# Print the execution time
print(f"Execution time: {execution_time} seconds")

# Check the length of json_list
print(f"Length of json_list: {len(json_list)}")
```

## Convert Json to DataFrame in order to perform data analysis

```python
col_names =  ['published_date','source_name','title','body','url']

df= pd.DataFrame(json_list,columns=col_names)
```

```python
df.shape
df.isnull().values.any()

#Chnaging date-time format 
df['published_date'] = pd.to_datetime(df['published_date'])
```

```python
df.sort_values(by=['published_date'],ascending=True)
df=df.reset_index(inplace=False)
df.head(100)
```
